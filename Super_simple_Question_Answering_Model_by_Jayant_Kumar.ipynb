## **Super simple Question Answering Model**

### **Loading Transformers and pipeline**

```python
from transformers import pipeline

qa_model = pipeline("question-answering", "timpal0l/mdeberta-v3-base-squad2")
```

### **Loading Context or string dataset**

```python
context = "My name is Tim and I live in Sweden . my second home is delhi"
```

```python
question = "what is my name?"
qa_model(question = question, context = context)
```

```python
question = "Where do I live?"
qa_model(question = question, context = context)
```

```python
question = "what is my second home?"
qa_model(question = question, context = context)
```
